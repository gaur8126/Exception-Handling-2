## Q.3 

```python
# 1.ZeroDivisionError: This error occurs when you try to divide a number by zero.
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error:", e)


# 2. OverflowError: This error occurs when a mathematical operation results in a value that exceeds the maximum representable value for a numeric type.

try:
    result = 10 ** 1000  # 10 raised to the power of 1000
except OverflowError as e:
    print("Error:", e)
```

## Q.4

```python
# 1. KeyError: This exception occurs when you try to access a key in a dictionary that does not exist.
my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['d']
except KeyError as e:
    print("Error:", e)

    
# 2. IndexError: This exception occurs when you try to access an index in a sequence (e.g., list, tuple, string) that is out of range.

my_list = [1, 2, 3]
try:
    value = my_list[3]
except IndexError as e:
    print("Error:", e)
```
